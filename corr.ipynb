```
from scipy import stats
```

```
coef,p_value = stats.pearsonr(df['Adj Close'],df['Daily Confirmed'])
```

```
print('Pearson’s Correlation Coefficient')
print('-'*40)
print('Correlation Test: %.3f' % coef)
# interpret the significance
alpha = 0.05
if p_value > alpha:
	print('H0: the two variables are independent. p=%.3f' % p_value)
else:
	print('H1: there is a dependency between the variables. p=%.3f' % p_value)
```

```
coef5,p_value5 = stats.spearmanr(df['Adj Close'],df['Daily Confirmed'])
```

```
print('Spearman’s Rank Correlation')
print('-'*40)
print('Spearmans correlation coefficient: %.3f' % coef5)
# interpret the significance
alpha = 0.05
if p_value > alpha:
	print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p_value5)
else:
	print('Samples are correlated (reject H0) p=%.3f' % p_value5)
```
